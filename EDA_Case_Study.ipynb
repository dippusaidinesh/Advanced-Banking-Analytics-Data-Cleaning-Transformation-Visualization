# Data Cleaning

####  Treating Duplicates

```python
df.duplicated().sum()   # conveys the duplicated data
```

```python
df = df.drop_duplicates()   # drops duplicates
```

#### Fixing DATE COLUMN

```python
df['Loan_Issue_Date'] = pd.to_datetime(df['Loan_Issue_Date'],errors = 'coerce')
```

#### Handling Missing Values

```python
df.isnull().sum()    # shows the null(missing) values in each column 
```

```python
df.head()
```

```python
# Treating Numeric Missing values
num_col = ['Annual_Income','Credit_Score','Loan_Amount','EMI']
for col in num_col:
    df[col] = df[col].fillna(df[col].median(), inplace = True)
```

```python
# Treating Categorical Data Missing values
df['Employment_Status'] = df['Employment_Status'].fillna(df['Employment_Status'].mode()[0],inplace = True)
df['Collateral_Type'] = df['Collateral_Type'].fillna(df['Collateral_Type'].mode()[0], inplace = True)
```

```python
# Treating  Date-time Missing values
df['Loan_Issue_Date'] = df['Loan_Issue_Date'].fillna(df['Loan_Issue_Date'].median(), inplace = True)
```

```python
df.isnull().sum() 
```

#### Treating Inconsistent Data

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\n🔹 {col} Value Counts:")
    print(df[col].value_counts())
```

```python
df['Gender'] = df['Gender'].replace({'M':'Male', 'F':'Female'})
df['Employment_Status'] = df['Employment_Status'].replace({'SELF employed':'Self-Employed'})
df['Default_Status'] = df['Default_Status'].replace({'yes':'Yes'})
df['Bank_Name'] = df['Bank_Name'].replace({'sbi':'SBI'})
```

#### Outliner Detection

```python
from feature_engine.outliers import Winsorizer

num = df.select_dtypes(include=['int64','float64']).columns.tolist()

exclude_cols = ['Customer_ID']
num = [col for col in num if col not in exclude_cols]

win = Winsorizer(capping_method='iqr', tail='both', variables=num)

df[num] = win.fit_transform(df[num])
```

```python
def UniNum(df,col):
    plt.figure(figsize=(6,4))
    plt.subplot(131)
    sns.boxplot(x = df[col],color='red')
    plt.subplot(132)
    sns.histplot(x = df[col],color='yellow')
    plt.subplot(133)
    sns.kdeplot(x = df[col],color='green',fill=True)
    plt.tight_layout()
    plt.show()
```

```python
lst = ['Annual_Income','Monthly_Income','Processing_Fee']
for col in lst:
    UniNum(df,col)
```
